```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
```

```python
model = ChatOpenAI(
    model="gpt-4o",  
    temperature=0
)
```

```python
# Note the strict system instruction
rag_template = ChatPromptTemplate.from_messages(
    [
        ("system","""your are a Legal Asistant.use the provide DOCUMENT CONTEXT to answer the question
                     If the answer is not in the context,say 'I Cannot find this in our database,please contact Admin'.
                     Do NOT use your own knowledge."""),
        ("user","DOCUMENT CONTEXT \n{context}\n\nQuestion:{question}")
    ]
)
```

```python
# Simulating a RAG retrieval step
context_data = "Section 4.2: Employees are entitled to 20 days of annual leave."
query= "i have apply 21 days annual leave.what should i do?"
```

```python
formatted_prompt = rag_template.invoke({
    "context":context_data,
    "question":query
})
```

```python
print(formatted_prompt.to_messages())
```

```python
response = model.invoke(formatted_prompt.to_messages())
```

```python
print(response)
```
